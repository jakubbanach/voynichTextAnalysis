```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from nltk.util import ngrams
from collections import Counter
from collections import defaultdict
import nltk
```

```python
with open('voynichFullTextClean.txt', 'r', encoding="utf8") as f:
    lines = f.readlines()
lines[0]
```

```python
x = lines[0].split(" ")
all_ngrams_voynich = []

for i in range(2,13):
    list_i_gram = {}
    NGRAMS = list(ngrams(x, i))
    print(f"{i}-GRAMY")
    for NGRAM in NGRAMS:
        if "" not in NGRAM:
            if NGRAM in list_i_gram:
                list_i_gram[NGRAM]+=1
            else:
                list_i_gram[NGRAM]=1
    sorted_NGRAMS = sorted(list_i_gram.items(), key=lambda x: x[1], reverse=True)

    df_grams = pd.DataFrame(sorted_NGRAMS, columns=['połączenia', 'wystąpienia'])
    df_grams.set_index('połączenia', inplace=True)
    df_grams.drop(df_grams[df_grams['wystąpienia'] < 2].index, inplace=True)

    for grams, number in sorted_NGRAMS:
        if number>1:
            print(f"{grams}: {number} wystąpienia")
    all_ngrams_voynich.append(df_grams)
```
