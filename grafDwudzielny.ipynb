```python
import pandas as pd
import numpy as np

with open('voynichFullTextClean.txt', 'r', encoding="utf8") as f:
    lines = f.readlines()

x = lines[0].split()
counter = {}
final = {}

for i in range(0, len(x)-1):
    current = x[i]
    next = x[i+1]
    if current not in counter:
        counter[current] = {}
        final[current] = 0
    
    if next not in counter[current]:
        counter[current][next] = 1
        final[current] += 1
    else:
        counter[current][next] += 1
    

df = pd.DataFrame(counter).fillna(0).astype(int)

final = pd.DataFrame.from_dict(final, orient='index', columns=['count']).sort_values(by=['count'], ascending=False)
```
